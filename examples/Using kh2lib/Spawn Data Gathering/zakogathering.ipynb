```python
from khbr.randomizer import KingdomHearts2
from kh2lib.kh2lib import kh2lib
import os, json, yaml
lib = kh2lib()
records = KingdomHearts2().enemy_records
enemies = set([e["obj_id"] for e in records.values() if e["type"] == "enemy"])
objlist = yaml.load(open("objlist.yml"))
objlist[1]
```

```python
objects = {}
for line in open("obj.md"):
    parts = [i.strip() for i in line.split("|")]
    objects[parts[1]] = parts[-2]
```

```python
obj_defs = {}
for o in objlist.values():
    tpe = o["ObjectType"]
    oid = o["ObjectId"]
    if tpe in ["ZAKO", "L_ZAKO"] and oid not in enemies:
        print("{} - {}".format(objects[str(oid)], tpe))
        name = objects[str(oid)]
        obj = {
            "aiparam": "",
            "animation": o["AnimationName"],
            "model": o["ModelName"],
            "name": name,
            "size": 0,
            "spawns": [],
            "type": "enemy",
            "ucm": hex(oid)[2:],
            "enmp": 999,
            "hp": 100
        }
        obj_defs[name] = obj
json.dump(obj_defs, open("enemies_extra.json", "w"), indent=4)
```
