```python
from kh2lib.kh2lib import kh2lib
import os, json
lib = kh2lib()
if not os.path.isdir("workspace"):
    os.mkdir("workspace")

msns = {}

msndir = os.path.join(lib.gamedir, "bars", "msn")
for msn in os.listdir(msndir):
    lfn = [i for i in os.listdir(os.path.join(msndir, msn)) if i.endswith(".list")][0]
    
    msnfn = os.path.join(lib.gamedir, "KH2", "msn", "jp", msn+".bar")
    
    subdata = lib.parsingengine.load_memdump(msnfn)
    
    starting_offset = int(lib.parsingengine.locate_file(os.path.join(msndir,msn,lfn)), 16)
    
    subdata_list = bytearray(open(os.path.join(msndir, msn, lfn), "rb").read())
    
    msns[msn] = {"list_offset": starting_offset, "bonus": subdata_list[0x0D]}
```

```python
json.dump(msns, open("msns.json", "w"))
```
